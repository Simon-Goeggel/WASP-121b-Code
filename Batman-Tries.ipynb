```python
import batman
import numpy as np
import matplotlib.pyplot as plt

params = batman.TransitParams()       #object to store transit parameters
params.t0 = 0.                        #time of inferior conjunction
params.per = 1.                       #orbital period
params.rp = 0.5                       #planet radius (in units of stellar radii)
params.a = 15.                        #semi-major axis (in units of stellar radii)
params.inc = 90.                      #orbital inclination (in degrees)
params.ecc = 0.                       #eccentricity
params.w = 90.                        #longitude of periastron (in degrees)
params.limb_dark = "nonlinear"        #limb darkening model
params.u = [0.5, 0.1, 0.1, -0.1]      #limb darkening coefficients [u1, u2, u3, u4]

t = np.linspace(-0.025, 0.025, 1000)  #times at which to calculate light curve
m = batman.TransitModel(params, t)    #initializes model
```

```python
flux = m.light_curve(params)                    #calculates light curve
```

```python
plt.plot(flux) #plotting the light curve and a grit pattern
plt.grid()
plt.show()
```

```python
params = batman.TransitParams()       #object to store transit parameters
params.t0 = 0.                        #time of inferior conjunction
params.per = 1.                       #orbital period
params.rp = 0.5                       #planet radius (in units of stellar radii)
params.a = 15.                        #semi-major axis (in units of stellar radii)
params.inc = 90.                      #orbital inclination (in degrees)
params.ecc = 0.                       #eccentricity
params.w = 90.                        #longitude of periastron (in degrees)
params.limb_dark = "nonlinear"        #limb darkening model
params.u = [0.5, 0.1, 0.1, -0.1]      #limb darkening coefficients [u1, u2, u3, u4]

t = np.linspace(-0.025, 0.025, 1000) #creating a time model where the values can be plotted
per_options = [0.25, 0.5,1,2,5] #creating a list of different options for the period

plt.figure()

    
for i in range(5):   #creating a for-loop to plot a transit light curve with every value for the period from the afore created list
        params.per = per_options[i]          
        
                    
        m = batman.TransitModel(params, t)        
        flux = m.light_curve(params)              
        plt.plot(t, flux, label =  per_options[i]   )
        
        plt.legend(loc='upper left')
plt.title("Orbital Period")        
plt.grid()
plt.xlabel('Time')
plt.ylabel('Relative Flux')
plt.show()
params.per = 1.  
```

```python
t = np.linspace(-0.025, 0.025, 1000)
rp_options = [0.15, 0.25, 0.4,  0.5, 0.75] # repeating the same process from the cell before but with the planets radius (in stellar radii)

plt.figure()


for i in range(5):  
        params.rp = rp_options[i]          
        
                    
        m = batman.TransitModel(params, t)        #initializes the model
        flux = m.light_curve(params)              #calculates light curve
        plt.plot(t, flux, label =  rp_options[i]   )
        
        plt.legend(loc='upper left')
plt.title("Planet Radius (in Units of Stellar Radii)")  
plt.xlabel('Time')
plt.ylabel('Relative Flux')
plt.grid()
plt.show()
params.rp = 0.5  
```

```python
ld_options = ["uniform", "linear", "quadratic", "nonlinear"] #calculating a light curve for all the diferent limb darkening options
ld_coefficients = [[], [0.3], [0.1, 0.3], [0.5, 0.1, 0.1, -0.1]]

plt.figure()


for i in range(4): 
        params.limb_dark = ld_options[i]          #specifies the LD profile
        
        params.u = ld_coefficients[i]             #updates LD coefficients
        m = batman.TransitModel(params, t)        #initializes the model
        flux = m.light_curve(params)              #calculates light curve
        plt.plot(t, flux, label = ld_options[i])
        
        plt.legend()
plt.title("Limb Darkening Parameters")        
plt.grid()
plt.xlabel('Time')
plt.ylabel('Relative Flux')
plt.show()
```
